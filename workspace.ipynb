```python
from scene.dataset_readers import readColmapSceneInfo, readSynthetic4RelightInfo
import random
import matplotlib.pyplot as plt
```

```python
def display_random_images_and_masks(scene_info, num_samples=10, figsize=(15, 8)):
    """
    Display random images and their corresponding masks from a dataset.
    
    Args:
        scene_info: SceneInfo object containing camera information
        num_samples: Number of random images to display (default: 10)
        figsize: Size of the figure (default: (15, 8))
        
    Returns:
        None (displays the images and masks)
    """
    # Combine train and test cameras
    all_cameras = scene_info.train_cameras + scene_info.test_cameras
    
    # Check if we have enough cameras
    if len(all_cameras) < num_samples:
        num_samples = len(all_cameras)
        print(f"Warning: Only {num_samples} cameras available.")
    
    # Select random cameras
    random_cameras = random.sample(all_cameras, num_samples)
    
    # Create a figure with subplots
    fig, axes = plt.subplots(num_samples, 2, figsize=figsize)
    
    # If only one sample, make axes 2D
    if num_samples == 1:
        axes = axes.reshape(1, -1)
    
    # Display each image and its mask
    for i, camera in enumerate(random_cameras):
        # Display the image
        axes[i, 0].imshow(camera.image)
        axes[i, 0].set_title(f"Image: {camera.image_name}")
        axes[i, 0].axis('off')
        
        # Display the mask
        axes[i, 1].imshow(camera.image_mask)
        axes[i, 1].set_title(f"Mask: {camera.image_name}")
        axes[i, 1].axis('off')
    
    plt.tight_layout()
    plt.show()
    
    return random_cameras  # Return the selected cameras for further inspection if needed
```

```python
dataset_path = "/userhome/30/alaksana/datasets/table/"  # Replace with your actual path

# Load the Colmap dataset
scene_info = readColmapSceneInfo(path=dataset_path, images="images", eval=False)
```

```python
random_cameras = display_random_images_and_masks(scene_info, num_samples=10, figsize=(15, 20))

# Optional: Print information about the first random camera
camera = random_cameras[0]
print(f"Camera ID: {camera.uid}")
print(f"Image dimensions: {camera.width} x {camera.height}")
print(f"Image name: {camera.image_name}")
```
